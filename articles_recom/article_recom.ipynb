```python

    import pickle as pk
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.metrics.pairwise import cosine_similarity
    import re
    from nltk.stem.snowball import SnowballStemmer
    import nltk
    stemmer = SnowballStemmer("english")
```

```python

#articles is a list of all Text
articles = ch['Text'].tolist()
articles[0] #an uncleaned article
```

```python
def clean_tokenize(document):
    document = re.sub('[^\w_\s-]', ' ',document)       #remove punctuation marks and other symbols
    tokens = nltk.word_tokenize(document)              #Tokenize sentences
    cleaned_article = ' '.join([stemmer.stem(item) for item in tokens])    #Stemming each token
    return cleaned_article
cleaned_articles = list (map(clean_tokenize, articles))
cleaned_articles[0] #a cleaned, tokenized and stemmed article 
```

```python
ch['cleaned_text'] = cleaned_articles
```

```python
#titles is a list of all titles
titles = ch['Title_y'].tolist()
titles[0] #an uncleaned article
```

```python
cleaned_titles = list (map(clean_tokenize, titles))
cleaned_titles[0] #a cleaned, tokenized and stemmed article 
```

```python
ch['cleaned_titles'] = cleaned_titles
```

```python
ch.drop(columns=['Text','Title_y'],inplace=True)
```

```python
ch.to_csv('cleaned_final_data.csv')
```

```python
# Articles index----- creating article id for all the titles
ai = pd.DataFrame(ch['cleaned_titles'].value_counts().index,columns={'cleaned_titles'})
```

```python
gh = pd.read_csv('data_with_index.csv')
```

```python
gh.drop(columns='Unnamed: 0',inplace=True)
```

```python
gh
```

```python
# articles recommended for different visitors(10 articles recommended)
dic={}
for i in gh['Visitor']:
    gh1 = gh[gh['Visitor']==i]
    ARTICLES_READ=gh1['text_index'].values
    #Get user representation in terms of words associated with read articles
    user_articles = ' '.join(cleaned_articles[p] for p in ARTICLES_READ)
    user_articles
    #Generate tfidf matrix model for entire corpus
    tfidf_matrix = TfidfVectorizer(stop_words='english', min_df=2)
    article_tfidf_matrix = tfidf_matrix.fit_transform(cleaned_articles)
    article_tfidf_matrix #tfidf vector of an article
    #Generate tfidf matrix model for read articles
    user_article_tfidf_vector = tfidf_matrix.transform([user_articles])
    user_article_tfidf_vector
    user_article_tfidf_vector.toarray()
    #  Calculate cosine similarity between user read articles and unread articles 
    articles_similarity_score=cosine_similarity(article_tfidf_matrix, user_article_tfidf_vector)
    recommended_articles_id = articles_similarity_score.flatten().argsort()[::-1]
    NUM_RECOMMENDED_ARTICLES=10
    #Remove read articles from recommendations
    final_recommended_articles_id = [article_id for article_id in recommended_articles_id 
                                 if article_id not in ARTICLES_READ ][:NUM_RECOMMENDED_ARTICLES]
    print('Titles of recommended articles for ', i)
    for j in final_recommended_articles_id:
        tex=ai[ai['article_id']==j]['cleaned_titles']
        #print('titles of recommended articles are:',j)
        print('Article id: ',j)
        print('Article title: ',tex)
        dic.update({j:tex})
    print(' ')
    print(' ')
    print('---------------')
    print(' ')
    print(' ')
    
```
